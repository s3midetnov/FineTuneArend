```python
import json
import random
import pathFile
from FineTuneLLM import datasetPreprocessing as pp
```

# Make random datasets for training, validation and testing

```python
# Parsed arend data
arend_data = pp.parse_concatenated_json(pathFile.dataset_maker_arend_data)

# Paths for the new dataset
path_train = pathFile.dataset_maker_train
path_valid = pathFile.dataset_maker_valid
path_test = pathFile.dataset_maker_test
path_user_prompts = pathFile.dataset_maker_user_prompts
path_user_queries = pathFile.dataset_maker_user_queries

# Sizes for random sampling for the new dataset
train_size = 30_000
valid_size = 1_000
test_size = 1_000
user_test_size = 100
```

```python
total_size = train_size + valid_size + test_size + user_test_size

total_sample = random.sample(arend_data, total_size)

train_sample = total_sample[0 : train_size]
with open(path_train, 'w', encoding='utf-8') as f:
    for entry in train_sample:
        f.write(json.dumps(pp.simplify_jsonl(entry)) + "\n")

valid_sample = total_sample[train_size : train_size  + valid_size]
with open(path_valid, 'w', encoding='utf-8') as f:
    for entry in valid_sample:
        f.write(json.dumps(pp.simplify_jsonl(entry)) + "\n")

test_sample = total_sample[train_size + valid_size : train_size + valid_size + test_size]
with open(path_test, 'w', encoding='utf-8') as f:
    for entry in test_sample:
        f.write(json.dumps(pp.simplify_jsonl(entry)) + "\n")

user_sample = total_sample[train_size + valid_size + test_size : train_size + valid_size + test_size + user_test_size]
with open(path_user_prompts, 'w', encoding='utf-8') as f:
    for entry in user_sample:
        f.write(json.dumps(pp.simplify_jsonl(entry)) + "\n")

with open(path_user_queries, 'w', encoding='utf-8') as f:
    for entry in user_sample:
        f.write(json.dumps(entry) + "\n")
```

# make user tests

```python
user_query_index = random.randint(0, user_test_size-1)

with open(path_user_queries, 'r', encoding='utf-8') as f:
    for line_number, line in enumerate(f, start=1):
        if line_number == user_query_index:
            user_q = json.loads(line)
            print(json.dumps(user_q, indent=2))
            break
```
