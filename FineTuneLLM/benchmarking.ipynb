```python
import pathFile
import json
from FineTuneLLM import ModelInference as mi
```

```python
path_user_prompts = pathFile.dataset_maker_user_prompts
path_user_queries = pathFile.dataset_maker_user_queries
```

```python
def simple_benchmark(i : int, filename : str) -> None:
    user_query_index = i
    with open(filename, 'w') as writefile:
        writefile.write(f"index = {user_query_index}\n\n")
        writefile.write("the query:")
        with open(path_user_queries, 'r', encoding='utf-8') as f:
            for line_number, line in enumerate(f, start=1):
                if line_number == user_query_index:
                    user_q = json.loads(line)
                    writefile.write(str(json.dumps(user_q, indent=2)) + "\n")

        writefile.write("model output: \n")
        with open(path_user_prompts, 'r', encoding='utf-8') as f:
            for line_number, line in enumerate(f, start=1):
                if line_number == user_query_index:
                    user_p = json.loads(line)
                    writefile.write(str(mi.call_model(str(user_p['prompt']))) + "\n")
```

```python
for i in range(100):
    simple_benchmark(i, f"benchmarks/benchmark_output{i}.txt")
```
