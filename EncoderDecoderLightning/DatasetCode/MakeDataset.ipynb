# Make random datasets for training, validation and testing

```python
# Parsed arend data
arend_data = pp.parse_concatenated_json("/Users/artem.semidetnov/Documents/Predictor/data/data.json")

# Paths for the new dataset
path_train = "/EncoderDecoderLightning/DatasetCode/dataset/train.json"
path_valid = "/EncoderDecoderLightning/DatasetCode/dataset/val.json"
path_test = "/EncoderDecoderLightning/DatasetCode/dataset/test.json"

# Sizes for random sampling for the new dataset
train_size = 30_000
valid_size = 1_000
test_size = 1_000
```

```python
total_size = train_size + valid_size + test_size

total_sample = random.sample(arend_data, total_size)

train_sample = total_sample[0 : train_size]
with open(path_train, 'w', encoding='utf-8') as f:
    for entry in train_sample:
        f.write(json.dumps(pp.jsonl_to_prompt_refactor0(entry)) + "\n")

valid_sample = total_sample[train_size : train_size  + valid_size]
with open(path_valid, 'w', encoding='utf-8') as f:
    for entry in valid_sample:
        f.write(json.dumps(pp.jsonl_to_prompt_refactor0(entry)) + "\n")

test_sample = total_sample[train_size + valid_size : train_size + valid_size + test_size]
with open(path_test, 'w', encoding='utf-8') as f:
    for entry in test_sample:
        f.write(json.dumps(pp.jsonl_to_prompt_refactor0(entry)) + "\n")
```

```python
from torch.utils.data import random_split

train_sample, valid_sample, test_sample = random_split(total_sample, [30_000, 1_000, 1_000])


with open(path_train, 'w', encoding='utf-8') as f:
    for entry in train_sample:
        f.write(json.dumps(entry) + "\n")

valid_sample = total_sample[train_size : train_size  + valid_size]
with open(path_valid, 'w', encoding='utf-8') as f:
    for entry in valid_sample:
        f.write(json.dumps(entry) + "\n")

test_sample = total_sample[train_size + valid_size : train_size + valid_size + test_size]
with open(path_test, 'w', encoding='utf-8') as f:
    for entry in test_sample:
        f.write(json.dumps(entry) + "\n")

!head /Users/artem.semidetnov/Documents/Predictor/EncoderDecoderLightning/Legacy/dataset/val.json 1
```
